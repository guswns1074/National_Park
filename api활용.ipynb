```python
#대전광역시 버스정류장
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

browser = "C:\\Users\CPB06GameN\Anaconda3\envs\chromedriver.exe"
driver = webdriver.Chrome(browser)
url = 'http://openapitraffic.daejeon.go.kr/api/rest/busRouteInfo/getStaionByRouteAll?serviceKey='
key = 'w1pN1f1ToPCuL6Lb1kVxIusAhz0tm5DHP%2B7y54wlkgBXSUIOrSzanJwd6hUPVD4CzoYOStNGFKlAtl6pW8LVwA%3D%3D'
page = '&reqPage='

bus_id = []
for i in range(0,110):
    driver.get(url+key+page+str(i))
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    a = soup.find_all('bus_node_id')
    b = soup.find_all('gps_lati')
    c = soup.find_all('gps_long')
    for j in range (0,len(a)):
        dic = {}
        dic['id'] = a[j].text
        dic['lat'] = b[j].text
        dic['lon'] = c[j].text
        bus_id.append(dic)
```

```python
bus_id
```

```python
new_bus_id = []
for i in range(0,len(bus_id)):
    if float(bus_id[i]['lat']) > 36.293291 and float(bus_id[i]['lat']) < 36.401253 and float(bus_id[i]['lon'])> 127.164302 and float(bus_id[i]['lon']) < 127.31:
        new_bus_id.append(bus_id[i])
```

```python
#광주광역시 버스정류장
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

browser = "C:\\Users\CPB06GameN\Anaconda3\envs\chromedriver.exe"
driver = webdriver.Chrome(browser)
url = 'http://api.gwangju.go.kr/xml/lineInfo?serviceKey='
key = 'w1pN1f1ToPCuL6Lb1kVxIusAhz0tm5DHP%2B7y54wlkgBXSUIOrSzanJwd6hUPVD4CzoYOStNGFKlAtl6pW8LVwA%3D%3D'
driver.get(url+key)
time.sleep(2)
html = driver.page_source
soup = BeautifulSoup(html,'lxml')
a = soup.find_all('line_id')
b = soup.find_all('first_run_time')
c = soup.find_all('last_run_time')
d = soup.find_all('run_interval')
del a[86],a[117]
del b[86]
del c[86]

gwangju_bus = []
for i in range(0,len(a)):
    dic = {}
    dic['bus_no'] = a[i].text
    dic['start_time'] = b[i].text
    dic['end_time'] = c[i].text
    dic['interval'] = d[i].text
    dic['count']= (((int(c[i].text[0:2])*60) + (int(c[i].text[3:6]))) - ((int(b[i].text[0:2])*60) + (int(b[i].text[3:6])))) // int(d[i].text)
    gwangju_bus.append(dic)
```
