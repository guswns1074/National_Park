```python
from selenium import webdriver
import pandas as pd
import time
```

```python
f = open("터미널명창.txt", 'r',encoding='utf8')
txt = f.read()
f.close()
```

```python
name_list = txt.split('\n')
```

```python
driver = webdriver.Chrome("C:\\Users\CPB06GameN\Anaconda3\envs\chromedriver.exe")
url = 'https://www.google.co.kr/maps/@45.2000678,17.2800773,5z?hl=ko'
driver.get(url)
```

```python
search_name = driver.find_element_by_css_selector('#searchboxinput')
search_button = driver.find_element_by_css_selector('#searchbox-searchbutton')
```

```python
terminal_locations = []
```

```python
for i in range(len(name_list)):
    search_name.clear()
    search_name.send_keys(name_list[i])
    search_button.click()
    time.sleep(5)
    
    url = driver.current_url
    split_url = url.split('/')
    split_url2 = split_url[6].split(',')
    point = split_url2[0:2]
    terminal_locations.append(point)
```

```python
len(terminal_locations)
```

```python
total = []
for i in range(0,len(name_list)):
    dic = {}
    dic['name'] = name_list[i]
    dic['location'] = terminal_locations[i]
    total.append(dic)
```

```python
terminal_locations.append([0,0])
```

```python
for i in range(0,len(total)):
    total[i]['location'][0] = total[i]['location'][0].replace('@','')
```
