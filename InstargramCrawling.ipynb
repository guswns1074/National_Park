```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import re     
import time
import json
from selenium.webdriver.common.keys import Keys
import urllib.parse
from konlpy.tag import Kkma

total=dict()
```

```python
search=input('검색어?')
search=urllib.parse.quote(search)
url='https://www.instagram.com/explore/tags/'+str(search)+'/'
```

```python
browser="C:\\Users\CPB06GameN\Anaconda3\envs\chromedriver.exe"
driver=webdriver.Chrome(browser)
driver.get(url)
```

```python
elem = driver.find_element_by_tag_name("body")
# alt 속성의 값을 담을 빈 리스트 선언
alt_list = set()

# 페이지 스크롤을 위해 임시 변수 선언
pagedowns = 1
# 스크롤을 20번 진행한다. #react-root > section > main > article > div:nth-child(3)
while len(alt_list) < 5000:
        # PAGE_DOWN(스크롤)에 따라서 결과 값이 달라진다.
        # 기본적으로 브라우저 조작을 통해 값을 얻어올 때는 실제 브라우저에 보이는 부분이어야 요소를 찾거나 특정 이벤트를 발생시킬 수 있다.
        elem.send_keys(Keys.PAGE_DOWN)
        if pagedowns%7==6:
            elem.send_keys(Keys.PAGE_UP)
        # 페이지 스크롤 타이밍을 맞추기 위해 sleep
        time.sleep(0.5) 
        # 브라우저에 보이는 모든 img 태그를 css 선택자 문법으로 찾는다.
        img = driver.find_elements_by_css_selector('div:nth-child(3) > div>div>div>a')
        print(len(img))
        # 위에서 선언한 alt_list 리스트에 alt 속성의 값을 할당한다.
        for i in img:
            alt_list.add(i.get_attribute('href'))
        pagedowns += 1
```
